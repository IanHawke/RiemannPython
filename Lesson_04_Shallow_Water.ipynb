# Shallow Water Equations

As a simple example of solving the Riemann problem for a nonlinear system we look at the *shallow water* equations. These are a simplification of the Navier-Stokes equations, reduced here to one spatial dimension $x$, which determine the height $h(x, t)$ of the water with respect to some reference location, and its velocity $u(x, t)$. In the simplest case (where the bed of the channel is flat, and the gravitational constant is renormalised to $1$) these can be written in the conservation law form

\begin{equation}
  \partial_t \begin{pmatrix} h \\ h u \end{pmatrix} + \partial_x \begin{pmatrix} hu \\ h u^2 + \tfrac{1}{2} h^2 \end{pmatrix} = {\bf 0}.
\end{equation}

The *conserved variables* ${\bf q} = (q_1, q_2)^T = (h, h u)^T$ are effectively the total mass and momentum of the fluid.

## Quasilinear form

As seen in the [theory lesson](Lesson_Theory.ipynb), to construct the solution we need the eigenvalues and eigenvectors of the Jacobian matrix. We can construct them directly, by first noting that, in terms of the conserved variables, 

\begin{equation}
  {\bf f} = \begin{pmatrix} f_1 \\ f_2 \end{pmatrix} = \begin{pmatrix} q_2 \\ \frac{q_2^2}{q_1} + \frac{q_1^2}{2} \end{pmatrix}.
\end{equation}

Therefore the Jacobian is

\begin{equation}
  \frac{\partial {\bf f}}{\partial {\bf q}} = \begin{pmatrix} 0 & 1 \\ -u^2 + h & 2 u \end{pmatrix}.
\end{equation}

The eigenvalues and eigenvectors follow immediately as

\begin{align}
  \lambda_{1} & = u - \sqrt{h}, & \lambda_{2} & = u + \sqrt{h}, \\
  {\bf r}_{1} & = \begin{pmatrix} 1 \\ u - \sqrt{h} \end{pmatrix}, & {\bf r}_{2} & = \begin{pmatrix} 1 \\ u + \sqrt{h} \end{pmatrix} .
\end{align}

Here we have followed the standard convention $\lambda_1 \le \lambda_2 \le \dots \le \lambda_N$.

An alternative approach that may be considerably easier to apply for more complex problems is to write down a different quasilinear form of the equation, which in this case is in terms of the *primitive variables* ${\bf w} = (h, u)^T$,

\begin{equation}
  \partial_t \begin{pmatrix} h \\ u \end{pmatrix} + \begin{pmatrix} u & h \\ 1 & u \end{pmatrix} \partial_x \begin{pmatrix} h \\ u \end{pmatrix} = {\bf 0}.
\end{equation}

The general form here would be written 

\begin{equation}
  \partial_t {\bf w} + B({\bf w}) \partial_x {\bf w} = {\bf 0}.
\end{equation}

It is straightforward to check that 

\begin{equation}
  B = \left( \frac{\partial {\bf q}}{\partial {\bf w}} \right)^{-1} \frac{\partial {\bf f}}{\partial {\bf w}} = \left( \frac{\partial {\bf q}}{\partial {\bf w}} \right)^{-1} \frac{\partial {\bf f}}{\partial {\bf q}} \left( \frac{\partial {\bf q}}{\partial {\bf w}} \right).
\end{equation}

Thus $B$ is *similar* to the Jacobian, so must have the same eigenvalues, which is straightforward to check. We also have that

\begin{align}
  B \left\{ \left( \frac{\partial {\bf q}}{\partial {\bf w}} \right)^{-1} {\bf r} \right\} = \lambda \left\{ \left( \frac{\partial {\bf q}}{\partial {\bf w}} \right)^{-1} {\bf r} \right\},
\end{align}

showing that the eigenvectors of the Jacobian can be straightforwardly found from the eigenvectors of $B$, which for the shallow water case are

\begin{align}
  {\bf \hat{r}}_1 &= \begin{pmatrix} -\sqrt{h} \\ 1 \end{pmatrix} & {\bf \hat{r}}_1 &= \begin{pmatrix} \sqrt{h} \\ 1 \end{pmatrix}.
\end{align}
